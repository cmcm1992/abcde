```python
import sqlite3
con = sqlite3.connect('news.db')
cur = con.cursor()
cur.execute("""
    SELECT content
    FROM table2
    ORDER BY id ASC
""")
corpus = [_[0] for _ in cur.fetchall()]
```

```python
def eojeol(sentence, N=2):
    tokens = word_tokenize(sentence) ## 토큰 갯수 가지고 n개 생성
    ngram = list()
    for i in range(len(tokens)-(N-1)):
        ngram.append(tuple(tokens[i:i+N]))
    return ngram
def umjeol(token, N=2):
    tokens=list(token)
    ngram = list()
    for i in range(len(tokens)-(N-1)):
        ngram.append("".join(tokens[i:i+N]))
#         ngram.append(tuple(tokens[i:i+N]))
    return ngram
```

```python
pattern4 = re.compile(r"\s{2,}")
collection = list()
for _ in corpus:
    collection.append(pattern4.sub(" ",pattern3.sub(" ",pattern2.sub(" ",pattern1.sub("",_)))))
```

```python
Vocabulary = list()
for _ in collection:
    Vocabulary.extend(_.split())
    Vocabulary.extend(eojeol(_))
    Vocabulary.extend(umjeol(_))
    Vocabulary.extend(kkma.morphs(_))
```

```python
Vocabulary = [_ for _ in list(set(Vocabulary)) if 1 < len(_) < 6]
len(Vocabulary)
```

```python
BoW =list(0 for _ in range(len(Vocabulary)))
for _ in collection[0].split() + eojeol(collection[0]) + umjeol(collection[0]) +kkma.morphs(collection[0]):
    if _ in Vocabulary:
        BoW[Vocabulary.index(_)] =1
```

```python
BoW =list()

for d in collection:
    docVector = list(0 for _ in range(len(Vocabulary)))
    for _ in d.split() + eojeol(d) + umjeol(d) +kkma.morphs(d):
        if _ in Vocabulary:
            docVector[Vocabulary.index(_)]=1
        BoW.append(docVector)    
```
